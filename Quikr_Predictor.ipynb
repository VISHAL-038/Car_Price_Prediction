Cleaning data


Working with Year

```python
#removing non numeric values
car=car[car['year'].str.isnumeric()]
```

```python
#data type change
car['year']=car['year'].astype(int)
```

Workiing with price


```python
#removing ask for price
car=car[car['Price']!='Ask For Price']
```

```python
#coversion of datatype
car['Price']=car['Price'].str.replace(',','').astype(int)
```

Working with kms_driven

```python
#removing kms
car['kms_driven']=car['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
#removing non numeric data
car=car[car['kms_driven'].str.isnumeric()]
```

```python
#datatype conversion
car['kms_driven']=car['kms_driven'].astype(int)
```

Working with Fuel type

```python
#removing nan
car=car[~car['fuel_type'].isna()]
```

Working with Name


```python
car['name']=car['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
#resetting the index
car=car.reset_index(drop=True)
```

Model


```python
x=car.drop(columns="Price")
y=car['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
#this encoder will convert categorical variables into a one-hot numeric array
#The fit method essentially learns the unique categories present in each of the specified columns ('name', 'company', and 'fuel_type') from the DataFrame x
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
#preprocessing steps to trasnform
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(Y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,Y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(Y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,Y_train)
y_pred=pipe.predict(X_test)
r2_score(Y_test,y_pred)
```
