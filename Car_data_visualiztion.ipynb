Visualize the data

```python
import matplotlib.pyplot as plt
```

on the basis on how diiferent feature effect the price

1. company vs price

```python
x = car['company']
y = car['Price']
plt.figure(figsize=(25, 5))
plt.scatter(x,y)
plt.xlabel('Company')
plt.ylabel('Price')
plt.show()
```

2. Year VS price

```python
x = car['year']
y = car['Price']

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Scatter Plot of Year vs Price')

# Show the plot
plt.show()
```

fuel vs price

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='fuel_type', y='Price', data=car)

# Add labels and title
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.title('Scatter Plot of Fuel Type vs Price')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

kms_driven vs price


```python
x = car['kms_driven']
y = car['Price']

# Create scatter plot
plt.figure(figsize=(6, 3))
plt.scatter(x, y)

# Add labels and title
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.title('Scatter Plot of Kilometers Driven vs Price')

# Show the plot
plt.show()
```

```python
model_counts = {}
for model in car['name']:
    if model not in model_counts:
        model_counts[model] = 0
    model_counts[model] += 1

# Prepare data for pie chart
labels = list(model_counts.keys())
sizes = list(model_counts.values())

# Create pie chart
plt.figure(figsize=(25, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Distribution of Car Models')

# Show the plot
plt.show()
```

```python
year_counts = {}
for year in car['year']:
    if year not in year_counts:
        year_counts[year] = 0
    year_counts[year] += 1

# Prepare data for pie chart
labels = list(year_counts.keys())
sizes = list(year_counts.values())

# Create pie chart
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Distribution of Cars by Year')

# Show the plot
plt.show()
```

```python
# Count the occurrences of each company
company_counts = {}
for company in car['company']:
    if company not in company_counts:
        company_counts[company] = 0
    company_counts[company] += 1

# Prepare data for pie chart
labels = list(company_counts.keys())
sizes = list(company_counts.values())

# Create pie chart
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Distribution of Cars by Company')

# Show the plot
plt.show()
```

```python
fuel_counts = {}
for fuel in car['fuel_type']:
    if fuel not in fuel_counts:
        fuel_counts[fuel] = 0
    fuel_counts[fuel] += 1

# Prepare data for pie chart
labels = list(fuel_counts.keys())
sizes = list(fuel_counts.values())

# Create pie chart
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Distribution of Cars by Fuel Type')

# Show the plot
plt.show()
```

```python
# Categorize prices into ranges
price_ranges = ['< 30000', '30000 - 49999', '50000 - 69999', '70000+']
price_counts = {range_label: 0 for range_label in price_ranges}

for price in car['Price']:
    if price < 30000:
        price_counts['< 30000'] += 1
    elif 30000 <= price < 50000:
        price_counts['30000 - 49999'] += 1
    elif 50000 <= price < 70000:
        price_counts['50000 - 69999'] += 1
    else:
        price_counts['70000+'] += 1

# Prepare data for pie chart
labels = list(price_counts.keys())
sizes = list(price_counts.values())

# Create pie chart
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Distribution of Car Prices')

# Show the plot
plt.show()
```

```python
kms_ranges = ['< 20000', '20000 - 39999', '40000 - 59999', '60000 - 79999', '80000+']
kms_counts = {range_label: 0 for range_label in kms_ranges}

for kms in car['kms_driven']:
    if kms < 20000:
        kms_counts['< 20000'] += 1
    elif 20000 <= kms < 40000:
        kms_counts['20000 - 39999'] += 1
    elif 40000 <= kms < 60000:
        kms_counts['40000 - 59999'] += 1
    elif 60000 <= kms < 80000:
        kms_counts['60000 - 79999'] += 1
    else:
        kms_counts['80000+'] += 1

# Prepare data for pie chart
labels = list(kms_counts.keys())
sizes = list(kms_counts.values())

# Create pie chart
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Distribution of Cars by Kilometers Driven')

# Show the plot
plt.show()
```
